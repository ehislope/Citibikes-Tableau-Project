# 2019 Data
## March Data

```python
d = pd.to_datetime(Mar2019_df['starttime'])
Mar2019_df['date'] = d.dt.date
Mar2019_df['time'] = d.dt.time
```

```python
# convert create month column and convert to month name
Mar2019_df['starttimemonth'] = pd.DatetimeIndex(Mar2019_df['date']).month
Mar2019_df['starttimemonth'] = Mar2019_df['starttimemonth'].apply(lambda x: calendar.month_abbr[x])
Mar2019_df.head()
```

```python
Mar2019_df.dtypes
```

```python
# calculate summary fields
# total trips per month
Trips_Mar2019 = Mar2019_df["tripduration"].count()
# Calculate total trips by subscriber and customer and percentage of each
Subtrips_Mar2019 = Mar2019_df.loc[Mar2019_df.usertype == "Subscriber", "usertype"].count()
percent_subtrips_Mar2019 = Subtrips_Mar2019/Trips_Mar2019*100
Custtrips_Mar2019 = Mar2019_df.loc[Mar2019_df.usertype == "Customer", "usertype"].count()
percent_custtrips_Mar2019 = Custtrips_Mar2019/Trips_Mar2019*100
# trips by gender (Gender (Zero=unknown; 1=male; 2=female))
maletrips_Mar2019 = Mar2019_df.loc[Mar2019_df.gender == 1, "gender"].count()
percent_maletrips_Mar2019 = maletrips_Mar2019/Trips_Mar2019*100
femaletrips_Mar2019 = Mar2019_df.loc[Mar2019_df.gender == 2, "gender"].count()
percent_femaletrips_Mar2019 = femaletrips_Mar2019/Trips_Mar2019*100
unknowntrips_Mar2019 = Mar2019_df.loc[Mar2019_df.gender == 0, "gender"].count()
percent_unknowntrips_Mar2019 = unknowntrips_Mar2019/Trips_Mar2019*100
# calculate total stations ids
start_station_id_count_Mar2019 = Mar2019_df["start station id"].nunique()
# Calculate total bike ids
bike_id_count_Mar2019=Mar2019_df["bikeid"].nunique()
bike_id_count_Mar2019
```

```python
# create monthly summary
Mar_2019_Summary = pd.DataFrame({
                     "Total Trips":[Trips_Mar2019],
                     "Subscriber Trips Count":[Subtrips_Mar2019],
                     "% Subscriber Trips":[percent_subtrips_Mar2019],
                     "Customer Trips Count":[Custtrips_Mar2019],
                     "% Customer Trips":[percent_custtrips_Mar2019],
                     "Trips by Males":[maletrips_Mar2019],
                     "% Trips by Males":[percent_maletrips_Mar2019],
                     "Trips by Females":[femaletrips_Mar2019],
                     "% Trips by Females":[percent_femaletrips_Mar2019],
                     "Trips by Unknown Gender":[unknowntrips_Mar2019],
                     "% Trips by Unknown Gender":[percent_unknowntrips_Mar2019],
                     "Start Station Count":[start_station_id_count_Mar2019],
                     "Bike ID Count":[bike_id_count_Mar2019]})
Mar_2019_Summary
```

```python
# add month column to summary
Mar_2019_Summary['Month'] = Mar2019_df['starttimemonth']
Mar_2019_Summary=Mar_2019_Summary[['Month','Total Trips','Subscriber Trips Count','% Subscriber Trips',
                                   'Customer Trips Count','% Customer Trips','Trips by Males', 
                                   '% Trips by Males','Trips by Females','% Trips by Females','Trips by Unknown Gender',
                                   '% Trips by Unknown Gender','Start Station Count','Bike ID Count']]
Mar_2019_Summary
```
